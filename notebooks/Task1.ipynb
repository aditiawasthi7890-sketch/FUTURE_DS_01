## Date Conversion

```python
monthly_sales = df.groupby(df["Order Date"].dt.to_period("M"))["Sales"].sum()
```

```python
monthly_sales
```

```python
df["Sales"].sum()
```

```python
import matplotlib.pyplot as plt
```

## Monthly Sales Trend

```python
monthly_sales.plot(figsize=(12,5))
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

### Insight

Sales fluctuate throughout the year, with noticeable peaks towards the end of each year, especially during November and December. This indicates seasonal demand and increased customer purchases during the festive period.

```python
print(monthly_sales)
```

```python
top_products = df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)

top_products
```

## Top 10 Selling Products

```python
plt.figure(figsize=(12,6))

top_products.plot(kind="bar")

plt.title("Top 10 Selling Products")
plt.xlabel("Product Name")
plt.ylabel("Sales")

plt.xticks(rotation=45)

plt.show()
```

### Insight

The Canon imageCLASS 2200 Advanced Copier generated the highest sales among all products. The company should ensure sufficient stock of high-performing products.

## Sales by Category

```python
category_sales = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)

category_sales
```

```python
df["Sales"].sum()
```

```python
plt.figure(figsize=(7,5))

category_sales.plot(kind="bar")

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")

plt.show()
```

### Insight

Technology generated the highest sales among all categories, followed by Furniture and Office Supplies. The company should continue investing in Technology products to maximize revenue.

## Sales by Region

```python
region_sales = df.groupby("Region")["Sales"].sum()

print(region_sales)
```

```python
region_sales.plot(kind="bar", figsize=(8,5))

plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")

plt.show()
```

### Insight

The West region generated the highest sales, while the South region recorded the lowest sales. The company should strengthen its marketing strategy in the South region to improve performance.

## Profit by Category

```python
category_profit = df.groupby("Category")["Profit"].sum()

print(category_profit)
```

```python
category_profit.plot(kind="bar", figsize=(8,5))

plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")

plt.show()
```

### Insight

Technology generated the highest profit. Although Furniture recorded good sales, its profit was comparatively low, indicating lower profit margins or higher discounts.




## Top 10 States by Sales

```python
state_sales = df.groupby("State")["Sales"].sum().sort_values(ascending=False).head(10)

print(state_sales)
```

```python
state_sales.plot(kind="bar", figsize=(10,5))

plt.title("Top 10 States by Sales")
plt.xlabel("State")
plt.ylabel("Sales")

plt.show()
```

### Insight

California achieved the highest sales, followed by New York and Texas. These states are the company's strongest markets.

## Top 10 Customers by Sales

```python
top_customers = df.groupby("Customer Name")["Sales"].sum().sort_values(ascending=False).head(10)

print(top_customers)
```

```python
top_customers.plot(kind="bar", figsize=(12,5))

plt.title("Top 10 Customers by Sales")
plt.xlabel("Customer Name")
plt.ylabel("Sales")

plt.xticks(rotation=45)

plt.show()
```

### Insight

Sean Miller is the highest revenue-generating customer. Retaining top customers through loyalty programs can help increase future sales.

## Discount vs Profit

```python
plt.figure(figsize=(8,5))

plt.scatter(df["Discount"], df["Profit"])

plt.title("Discount vs Profit")
plt.xlabel("Discount")
plt.ylabel("Profit")

plt.show()
```

### Insight

Higher discounts generally reduce profit. Orders with very high discounts often result in low or negative profits. The company should avoid excessive discounts to maintain profitability.
